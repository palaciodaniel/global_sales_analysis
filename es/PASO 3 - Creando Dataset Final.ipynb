```python
import pandas as pd
```

```python
# Cargando dataset de "Sales 2012"

sales2012 = pd.read_excel("cleaned_sales2012.xlsx", index_col = 0)

sales2012.sort_values("Date", axis = 0, ascending = True, inplace = True)

sales2012.reset_index(drop = True, inplace = True)

sales2012.head()
```

```python
sales2012.tail()
```

```python
sales2012.info()
```

```python
# Cargando dataset de "Sales 2013-14"

sales201314 = pd.read_excel("cleaned_sales201314.xlsx", index_col = 0)

sales201314.sort_values("Date", axis = 0, ascending = True, inplace = True)

sales201314.reset_index(drop = True, inplace = True)

sales201314.head()
```

```python
sales201314.tail()
```

```python
sales201314.info()
```

```python
# Concatenando ambos datasets en uno solo

sales = pd.concat([sales2012, sales201314], ignore_index = True)
sales.head()
```

```python
sales.tail()
```

```python
sales.info()
```

```python
del sales2012
del sales201314
```

```python
# Cargando dataset de clientes y ciudades

customer_city = pd.read_excel("cleaned_customer_city.xlsx", index_col = 0)

customer_city.sort_values("Region", axis = 0, ascending = True, inplace = True)

customer_city.reset_index(drop = True, inplace = True)

customer_city.head()
```

```python
customer_city.tail()
```

```python
# Como "Region" sólo tiene seis valores posibles, conviene cambiarlos al tipo "category" 
# para optimizar el rendimiento

customer_city["Region"] = customer_city["Region"].astype("category")
```

El siguiente punto fue crítico. Al unir este dataset con el de "Sales", de manera misteriosa el número de observaciones aumentaba, sin importar el tipo de join ("inner", "left", etc) que se usara. 

Tras [investigar debidamente](https://stackoverflow.com/questions/37095161/number-of-rows-changes-even-after-pandas-merge-with-left-option), se descubrió que hay un duplicado de ID en la columna "Customer Number", ya que la misma empresa posee una sucursal en Malmo (Suecia) y otra en Oslo (Noruega), pero el dataset las agrupa bajo un mismo "Customer Number", cuando lo ideal hubiera sido un número distinto para cada ciudad.

Por lo dicho, al simplemente usar "duplicated" estas dos observaciones eran pasadas por alto, ya que no eran consideradas un duplicado por tener ciudades distintas, y el inconveniente pudo ser detectado posteriormente (y a tiempo) gracias a usar ".info()" como sanity check.

```python
customer_city[customer_city.duplicated(subset = "Customer Number")]
```

```python
customer_city[customer_city["Customer Number"] == 10021911]
```

La consecuencia de esto es que al unir ambos datasets, los datos numéricos de las observaciones de "Sales" quedaban multiplicados por dos al ser aplicados a cada una de las ciudades, corrompiendo la estructura del dataset.

De este modo, una de las ciudades debe ser eliminada. Esto no ocasionará pérdida de información, como ya se indicó, porque de hecho se evita que la misma información quede repetida.

```python
customer_city.drop_duplicates(subset = "Customer Number", keep = 'first', inplace = True)
```

```python
customer_city.reset_index(drop = True, inplace = True)

customer_city.head(2)
```

```python
customer_city.tail(2)
```

```python
customer_city.info()
```

```python
# Uniendo el dataset de "Sales" con el clientes y ciudades

sales_final = pd.merge(sales, customer_city, how = "inner", on = "Customer Number")
sales_final.head()
```

```python
del sales
del customer_city
```

```python
sales_final.tail()
```

```python
sales_final.info()
```

```python
# La columna "Customer Number" se volvió innecesaria para los fines de este análisis.

sales_final.drop("Customer Number", axis = 1, inplace = True)
sales_final.head()
```

```python
# Agregar el dataset de clientes y ciudades alteró el orden de las fechas.

sales_final.sort_values("Date", axis = 0, ascending = True, inplace = True)

sales_final.reset_index(drop = True, inplace = True)

sales_final.tail(20)
```

```python
sales_final.info()
```

```python
# Guardar dataset final como XLSX

sales_final.to_excel("sales_final.xlsx", 
                       sheet_name = "Sales - Final Dataset", 
                       columns = sales_final.columns)
```

```python
# Guardar dataset final como CSV

sales_final.to_csv("sales_final.csv")
```
