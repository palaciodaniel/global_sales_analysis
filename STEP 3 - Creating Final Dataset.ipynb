```python
import pandas as pd
```

```python
# Loading "Sales 2012" dataset

sales2012 = pd.read_excel("cleaned_sales2012.xlsx", index_col = 0)

sales2012.sort_values("Date", axis = 0, ascending = True, inplace = True)

sales2012.reset_index(drop = True, inplace = True)

sales2012.head()
```

```python
sales2012.tail()
```

```python
sales2012.info()
```

```python
# Loading "Sales 2013-14" dataset

sales201314 = pd.read_excel("cleaned_sales201314.xlsx", index_col = 0)

sales201314.sort_values("Date", axis = 0, ascending = True, inplace = True)

sales201314.reset_index(drop = True, inplace = True)

sales201314.head()
```

```python
sales201314.tail()
```

```python
sales201314.info()
```

```python
# Concatenating both datasets into a single one

sales = pd.concat([sales2012, sales201314], ignore_index = True)
sales.head()
```

```python
sales.tail()
```

```python
sales.info()
```

```python
del sales2012
del sales201314
```

```python
# Loading dataset about clients and cities

customer_city = pd.read_excel("cleaned_customer_city.xlsx", index_col = 0)

customer_city.sort_values("Region", axis = 0, ascending = True, inplace = True)

customer_city.reset_index(drop = True, inplace = True)

customer_city.head()
```

```python
customer_city.tail()
```

```python
# Since "Region" column only has six possible values, it is convenient to change the column type to "category" 
# to optimize performance

customer_city["Region"] = customer_city["Region"].astype("category")
```

The next point was critical. When joining this dataset with the "Sales" one, misteriously the number of observations increased, no matter the "join" type ("inner", "left", etc) that was applied.

After [properly investigating](https://stackoverflow.com/questions/37095161/number-of-rows-changes-even-after-pandas-merge-with-left-option) the issue, it was discovered that there is an ID duplicate in the column "Customer Number", because the same company has a subsidiary in Malmo (Sweden) and another one in Oslo (Norway), but the dataset was grouping them under a similar "Customer Number", when the ideal classification should have been a distinct number for every city.

Therefore, when simply using the "duplicated" command, these two observations were overlooked, given that they were not considered duplicates due to having different cities, and the inconvenience could be detected later (and in time) thanks to using ".info()" as a sanity check.

```python
customer_city[customer_city.duplicated(subset = "Customer Number")]
```

```python
customer_city[customer_city["Customer Number"] == 10021911]
```

The consequence of this was that when joining both datasets, the numerical data from the "Sales" observations were multiplied by two due to being applied to each of the cities, corrupting in this way the dataset structure.

Thus, one of the cities must be deleted in order to solve the error. This will not cause information loss, as already indicated, because it actually prevents information from being repeated.

```python
customer_city.drop_duplicates(subset = "Customer Number", keep = 'first', inplace = True)
```

```python
customer_city.reset_index(drop = True, inplace = True)

customer_city.head(2)
```

```python
customer_city.tail(2)
```

```python
customer_city.info()
```

```python
# Joining "Sales" dataset with the one about clients and cities

sales_final = pd.merge(sales, customer_city, how = "inner", on = "Customer Number")
sales_final.head()
```

```python
del sales
del customer_city
```

```python
sales_final.tail()
```

```python
sales_final.info()
```

```python
# Column "Customer Number" became unnecessary for the objectives of this analysis.

sales_final.drop("Customer Number", axis = 1, inplace = True)
sales_final.head()
```

```python
# Adding the dataset about clients and cities altered the order of dates.

sales_final.sort_values("Date", axis = 0, ascending = True, inplace = True)

sales_final.reset_index(drop = True, inplace = True)

sales_final.tail(20)
```

```python
sales_final.info()
```

```python
# Saving the final dataset as a XLSX file

sales_final.to_excel("sales_final.xlsx", 
                       sheet_name = "Sales - Final Dataset", 
                       columns = sales_final.columns)
```

```python
# Saving the final dataset as a CSV file

sales_final.to_csv("sales_final.csv")
```
